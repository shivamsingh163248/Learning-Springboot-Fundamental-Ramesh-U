```python
totalNull = titanic.isnull().sum().sort_values(ascending=False)
#find the percentage
percent1 = titanic.isnull().sum()/titanic.isnull().count()*100
#again creating the round 
percent2 = round(percent1 , 1).sort_values(ascending=False)

#creating the 
data = pd.concat([totalNull , percent2] , axis=1 , keys=['total ' , '%'])
print(data)
```
